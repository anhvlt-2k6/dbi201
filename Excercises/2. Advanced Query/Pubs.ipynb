```sql
SELECT pub_name, COUNT(titles.title_id) as titlesCount
    FROM publishers, titles
    WHERE publishers.pub_id = titles.pub_id
    GROUP BY pub_name;
```

```sql
SELECT stor_name, SUM(sales.qty) AS totalBooks
    FROM stores, sales
    WHERE stores.stor_id = sales.stor_id
    GROUP BY stor_name;
```

```sql
SELECT stor_name, SUM(sales.qty * CONVERT(INT, titles.price)) AS totalPrice
    FROM stores, sales, titles
    WHERE stores.stor_id = sales.stor_id AND sales.title_id = titles.title_id
    GROUP BY stor_name
    ORDER BY stor_name;
```

```sql
SELECT TOP(4) stor_name, SUM(sales.qty * CONVERT(INT, titles.price)) AS totalPrice
    FROM stores, sales, titles
    WHERE stores.stor_id = sales.stor_id AND sales.title_id = titles.title_id
    GROUP BY stor_name
    ORDER BY totalPrice DESC;
```
