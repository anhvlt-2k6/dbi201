## Get destination-pairs

From table `Routes`, get detailed information about the departure-arrival airport pairs available

Print out tables include **Departure Airport**, **Arrival Airport**, and **Distance**


```sql
SELECT dep.ApName AS 'Departure Airport', arr.ApName AS 'Arrival Airport', r.Distance
    FROM Routes AS r
        LEFT JOIN Airports dep ON r.DepartICAO = dep.ApID
        LEFT JOIN Airports arr ON r.ArrICAO = arr.ApID
```

## Get flight information

From the view `AirlinesPerDestination`, get the friendly departure, arrivial airport, flight number, and airline name.


```sql
SELECT a.FlightNum AS 'Flight Number', a.AirlineName, dep.ApName AS 'Departure Airport', arr.ApName AS 'Arrival Airport'
    FROM AirlinesPerDestination AS a
        LEFT JOIN Airports AS dep ON a.DepartICAO = dep.ApID
        LEFT JOIN Airports AS arr ON a.ArrICAO = arr.ApID
```
