## Get the customer information

From `Passengers` table, get the `PassengerID` from any part of the name, email, and passport.


```sql
DECLARE @name NVARCHAR(MAX);
DECLARE @mail NVARCHAR(255);
DECLARE @passport NVARCHAR(255);

SET @name = 'Charles';
SET @mail = 'CharlesDCartagena@rhyta.com';
SET @passport = '435678564';

SELECT p.PassengerID
    FROM Passengers AS p
    WHERE (FirstName LIKE '%' + @name + '%' OR LastName LIKE '%' + @name + '%')
        AND Email = @mail
        AND PassportNum = @passport
```

## Get Seat information

From `AvailSeatPerAirlines` view, get `SeatID` based on range of Price, range of DepartTime, Departure and Arrival Airport (in ICAO).

Print out the seat ID and seat price. Order in price


```sql
DECLARE @upperPrice MONEY = 100;
DECLARE @lowerPrice MONEY = 0;
DECLARE @lowerDepart DATETIME2 = '2025-11-04 00:00:00';
DECLARE @upperDepart DATETIME2 = '2027-11-02 00:00:00';
DECLARE @depICAO VARCHAR(4) = 'KBOS';
DECLARE @arrICAO VARCHAR(4) = 'EHAM';

IF @lowerPrice > @upperPrice
BEGIN
    DECLARE @tmpPrice MONEY = @lowerPrice;
    SET @lowerPrice = @upperPrice;
    SET @upperPrice = @tmpPrice;
END

IF @lowerDepart > @upperDepart
BEGIN
    DECLARE @tmpDT DATETIME2 = @lowerDepart;
    SET @lowerDepart = @upperDepart;
    SET @upperDepart = @tmpDT;
END

SELECT
    a.SeatID,
    a.Price,
    a.DepartTime,
    a.DepartICAO,
    a.ArrICAO
FROM AvailSeatPerAirlines AS a
WHERE a.Price BETWEEN @lowerPrice AND @upperPrice
  AND a.DepartTime BETWEEN @lowerDepart AND @upperDepart
  AND a.DepartICAO = @depICAO
  AND a.ArrICAO = @arrICAO
ORDER BY a.Price ASC;
```
